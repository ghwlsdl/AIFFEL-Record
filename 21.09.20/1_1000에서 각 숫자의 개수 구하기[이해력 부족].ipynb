```python
Q. 1~1000에서 각 숫자의 개수 구하기

예로 10 ~ 15 까지의 각 숫자의 개수를 구해보자

10 = 1, 0
11 = 1, 1
12 = 1, 2
13 = 1, 3
14 = 1, 4
15 = 1, 5

그러므로 이 경우의 답은 0:1개, 1:7개, 2:1개, 3:1개, 4:1개, 5:1개
```

```python
# 가영님 아이디어
num = 10
for i in map(int,str(num)):
    print(i)
```

```python
number=[] # 리스트화 한 숫자
for i in range(1,1000+1): # 범위 지정
    number += str(i) # 문자 별 숫자 더하기
print(number)
```

```python
number=[] # 리스트화 한 숫자
for i in range(1,1000+1): # 범위 지정
    number += str(i) # 문자 별 숫자 더하기
for n in range(0,9+1): # 위의 더한 숫자를 0 ~ 9 사이의 숫자로 정렬
    print (str(n),':개 ', number.count(str(n)))
```
