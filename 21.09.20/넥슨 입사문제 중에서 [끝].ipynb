```python
# Q. 어떤 자연수 n이 있을 때,

# d(n)을 n의 각 자릿수 숫자들과 n 자신을 더한 숫자라고 정의하자.

# 예를 들어, d(91) = 9 + 1 + 91 = 101

# 이 때, n을 d(n)의 제네레이터(generator)라고 한다.

# 위의 예에서 91은 101의 제네레이터이다.

# 어떤 숫자들은 하나 이상의 제네레이터를 가지고 있는데,

# 101의 제네레이터는 91 뿐 아니라 100도 있다.

# 그런데 반대로, 제네레이터가 없는 숫자들도 있으며,

# 이런 숫자를 인도의 수학자 Kaprekar가 셀프 넘버(self-number)라 이름 붙였다.

# 예를 들어 1,3,5,7,9,20,31 은 셀프 넘버 들이다.

# 1 이상이고 5000 보다 작은 모든 셀프 넘버들의 합을 구하라.
```

```python
def self_number():   # self_number 정의
    arrange=list()   # list 정의
    result=None      # return 정의
    
    for n in range(1, 5000+1):
        
        if n < 10: # 10 이하 자연수에 대해서 d(n) 정의
            result = n+n
            arrange.append(result)
            
        elif n < 100: # 100 이하 자연수에 대해서 d(n) 정의
            result = n+(n//10)+(n%10)
            arrange.append(result)
            
        elif n < 1000: # 100 이하 자연수에 대해서 d(n) 정의
            result = n+(n//100)+((n%100)//10)+((n%100)%10)
            arrange.append(result)
            
        elif n < 5000: # 5000 이하 자연수에 대해서 d(n) 정의
            result = n+(n//1000)+((n%1000)//100)+(((n%1000)%100)//10)+(((n%1000)%100)%10) 
            if result <= 5000:
                arrange.append(result)
                
    arrange.sort() # 오름차순 정렬이 진행
    
    arrange1 = [n for n in range(1,5000+1)]
    
    notSelf = [item for item in arrange1 if item not in arrange]
    
    #for each in notSelf: # 주석처리 한 이유? 결과적으로 숫자를 나열하는 것이 아니라, 전체 (arrange1) 에서
    #    print(each)      #  notSelf 한 값을 뺀 Self_number 함수의 합을 원한 것이기 때문
    return notSelf
 
self_number()
```
